### Scatter plot and Correlation Strength

```python
x = data['daily']
y = data['sunday']
plt.scatter(data['daily'], data['sunday'])
plt.xlim(0, max(x) + 100)
plt.ylim(0, max(y) + 100)
plt.show()
```

```python
data['daily'].corr(data['sunday'])
```

```python
data[['daily','sunday']].corr()
```

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data).fit()
```

```python
model1.summary()
```

```python
x = data["daily"].values
y = data["sunday"].values
plt.scatter(x,y, color = "m", marker = "o" , s = 30)
b0 = 13.84
b1 = 1.33

y_hat = b0 + b1*x

plt.plot(x, y_hat, color = "g")

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
sns.regplot(x="daily", y="sunday", data=data)
plt.xlim([0,1250])
plt.show()
```

### Predict for new data points(test data)

```python
newdata=pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
#predict on all given training data
pred = model1.predict(data["daily"])
pred
```

```python
data["y_hat"] = pred
data
```

```python
data["residuals"]=data["sunday"]-data["y_hat"]
data
```

```python
mse = np.mean((data["daily"]-data["y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
mae = np.mean(np.abs(data["daily"]-data["y_hat"]))
mae
```

## Checking the model residuals scatter plot
 - check the scatter plot for residuals Vs Y-hat
 - check the Q-Q plot for residuals distribution
 - check the kde hist plot for residulas

```python
plt.scatter(data["y_hat"],data["residuals"])
plt.xlabel("Y_hat")
plt.ylabel("Residuals")
```

### Observations:
 - There appears to be no trend and the residuals are randomly placed round the zero error line
 - Hence the assumption of homoscedasticty(constant variance in resduals) is satisfied

```python
import statsmodels.api as sm
sm.qqplot(data["residuals"], line='45', fit=True)
plt.show()
```

```python
sns.histplot(data["residuals"], kde=True)
```

### Observations
 - The data points are seen to closely follow the reference line of normality
 - Hence the residuals are approximately normally distributed as also can be seen from the kde distribution
 - All the assumptions of simple linear regression are verified and found to be valid in this case
