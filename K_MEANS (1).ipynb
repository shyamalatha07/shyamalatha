```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
```

## Clustering-Divide the universities in to groups(clusters)

```python
uni=pd.read_csv("Universities.csv")
uni
```

```python
uni.info()
```

```python
uni.describe()
```

```python
uni1=uni.iloc[:,1:]
```

```python
uni1
```

```python
cols=uni1.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_uni_df=pd.DataFrame(scaler.fit_transform(uni1),columns=cols)
scaled_uni_df
```

```python
from sklearn.cluster import KMeans
clusters_new=KMeans(3,random_state=0)
clusters_new.fit(scaled_uni_df)
```

```python
clusters_new.labels_
```

```python
set(clusters_new.labels_)
```

```python
uni['clusterid_new']=clusters_new.labels_
```

```python
uni
```

```python
uni.sort_values(by="clusterid_new")
```

```python
uni.iloc[:,1:].groupby("clusterid_new").mean()
```

## Observations
- cluster2 appears to be the top rated universities cluster as the cut off score,top 10,sf ratio parameter mean values are highest
- cluster 1 appears to occupy the middle level rated universities
- cluster 0 comes as the lower level rated universities

```python
uni[uni['clusterid_new']==0]
```

```python
w=[]
for i in range(1,20):
    Kmeans=KMeans(n_clusters=i,random_state=0)
    Kmeans.fit(scaled_uni_df)
    w.append(Kmeans.inertia_)
print(w)
plt.plot(range(1,20),w)
plt.title("Elbow Method")
plt.xlabel("no.of clusters")
plt.ylabel("wcss")
plt.show()
```
